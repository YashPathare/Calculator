```python
import yfinance as yf
import pandas as pd

def fetch_stock_data(ticker="AAPL", period="1mo", interval="1d"):
    """
    Fetch historical stock data for a given ticker.
    :param ticker: Stock symbol (e.g., AAPL, TSLA)
    :param period: Data period (e.g., "1mo", "6mo", "1y", "5y", "max")
    :param interval: Interval (e.g., "1d", "1h", "5m")
    :return: DataFrame with stock price history
    """
    stock = yf.Ticker(ticker)
    data = stock.history(period=period, interval=interval)
    data.reset_index(inplace=True)  # Reset index for better handling
    return data

# Fetch Apple (AAPL) stock data for the last 1 month
stock_data = fetch_stock_data("AAPL", "1mo", "1d")
print(stock_data.head())  # Display the first few rows
```

```python
import requests

NEWS_API_KEY = "80972d7a8f134aefbd11fba3c9264a84"

def fetch_financial_news(query="stock market", page_size=10):
    """
    Fetch latest financial news using NewsAPI.
    :param query: Keyword for news search
    :param page_size: Number of articles to fetch
    :return: List of news articles
    """
    url = f"https://newsapi.org/v2/everything?q={query}&language=en&apiKey={NEWS_API_KEY}&pageSize={page_size}"
    response = requests.get(url).json()
    articles = response.get("articles", [])

    news_data = []
    for article in articles:
        news_data.append({
            "title": article["title"],
            "description": article["description"],
            "url": article["url"],
            "published_at": article["publishedAt"]
        })

    return pd.DataFrame(news_data)

# Fetch latest financial news
news_df = fetch_financial_news()
print(news_df.head())  # Display first few news articles
```

```python
from transformers import pipeline

# Load FinBERT model
sentiment_pipeline = pipeline("text-classification", model="ProsusAI/finbert")

def analyze_sentiment(news_df):
    """
    Perform sentiment analysis on news headlines using FinBERT.
    :param news_df: DataFrame containing news articles
    :return: DataFrame with sentiment scores
    """
    news_df["sentiment"] = news_df["title"].apply(lambda x: sentiment_pipeline(x)[0]["label"])
    return news_df

# Analyze sentiment of financial news
news_df = analyze_sentiment(news_df)
print(news_df[["title", "sentiment"]])
```

```python
def preprocess_stock_data(stock_df):
    """
    Clean and preprocess stock market dataset.
    :param stock_df: Raw stock market DataFrame
    :return: Cleaned and normalized DataFrame
    """
    stock_df = stock_df.dropna()  # Remove missing values
    stock_df["Date"] = pd.to_datetime(stock_df["Date"])  # Convert date column
    stock_df.set_index("Date", inplace=True)  # Set Date as index

    # Normalize using Min-Max Scaling
    stock_df["Open"] = (stock_df["Open"] - stock_df["Open"].min()) / (stock_df["Open"].max() - stock_df["Open"].min())
    stock_df["Close"] = (stock_df["Close"] - stock_df["Close"].min()) / (stock_df["Close"].max() - stock_df["Close"].min())

    return stock_df

# Apply cleaning to stock data
stock_data = preprocess_stock_data(stock_data)
print(stock_data.head())
```
